### What is boxplot?
- Boxplot - It is used to see the distribution of numerical data and identify key stats like maximun and minimum values, identify outliers, understanding how data is distributed and can compare the distribution of data across different categories (or) variables.
- Synatx - sns.boxplot( data=None, x=None, y=None, hue=None, order=None, hue_order=None, orient=None, color=None, palette=Npone, saturation=0.75, fill=True, dodge='auto', width=0.8, gap-0, whis=1.5, linecolor='auto', linewidth=None, fliersize=None, hue_norm=None, native_scale=False, log_scale=None, legend='auto', ax=None, **kwargs)

### Import all the required LIBRARIES


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Import dataframe

```python
mart = pd.read_excel("supermarket_sales.xlsx")
mart.head()
```

### Create a basic BOX plot on one NUMERIC variable

```python
sns.boxplot(y='Total',data=mart, width=0.2)
```

### Create a basic BOX plot on one numeric variable by a CATEGORICAL variable

```python
sns.boxplot(x='Payment', y='Total', data=mart)
```

### Create a basic BOX plot on one numeric variable by TWO CATEGORICAL variable using HUE attribute

```python
sns.boxplot(x='Payment', y='Total', hue='Gender', data=mart)
```

### Add MEAN marker in the box plot using Showmeans attribute and change its style using meanprops

```python
sns.boxplot(x='Payment',y='Total',hue='Gender',data=mart,showmeans=True,
            meanprops={
                'marker':'o',
                'markerfacecolor':'white',
                'markersize':'10',
                'markeredgecolor':'black' }
           )
```

### Make HORIZONTAL box plot

```python
sns.boxplot(x='Total',y='Payment',hue='Gender',data=mart,showmeans=True,
            meanprops={
                'marker':'o',
                'markerfacecolor':'white',
                'markersize':'5',
                'markeredgecolor':'black' }
           )
```

### Change PALETTE, LINE WIDTH etc..

```python
sns.boxplot(x='Total',y='Payment',hue='Gender',data=mart,showmeans=True,meanprops={"marker":"o"
                                                                                  ,"markerfacecolor":"white"
                                                                                  ,"markersize":"5"
                                                                                  ,"markeredgecolor":"black"}
           ,palette='CMRmap'
           ,linewidth=5)
```

### Create box plot for EACH OF THE NUMERIC VARIABLE in the dataframe

```python
plt.figure(figsize=(15,5))

sns.boxplot(data=mart)
```
