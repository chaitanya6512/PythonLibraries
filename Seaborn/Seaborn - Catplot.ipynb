### What is Catplot?
- This function provides access to several axes-level functions that show the relationship between a numerical and one-more categorical variables using one of several visual representations.
- Syntax - sns.catplot(*, x=None, y=None, hue=None, data=None, row=None, col=None, kind='strip', color=None, palette=None, **kwargs)
- The kind parameter selects underlying axes-level function to use :
- 1. Categorical scatterplots -
     - stripplot() - with kind="strip"; Default.
     - swarmplot() - with kind="swarm"
  2. Categorical distribution plots -
     - boxplot() - with kind="box".
     - violinplot() - with kind="violin".
     - boxenplot() - with kind="boxen".
  3. Categirical estimate plots -
     - pointplot() - with kind="point".
     - barplot() - with kind="bar".
     - countplot() - with kind="count".
