### What is KDE?
- KDE - Kernel Density Estimate(KDE).
- It offers detailed view of the probability density of continous variables.
- Syntax - sns.kdeplot(data, *, x=None, y=None, hue=None, weights=Non, palette=None, hue_order=None, hue_norm=None, color=None, fill=None, multiple='layer', common_norm=True, common_grid=False, cumulative=False, bw_method='scott', bw_adjust=1, warn_singular=True, log_sclae=None, levels=10, thresh=0.05, gridsize=200, cut=3, clip=None, legend=True, cbar=False, cbar_ax=None, cbar_kws=None, ax=None, **kwargs)

### Imported required libraries.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as sts
```

```python
pd.set_option('display.max_rows',None)
```

### Import Dataset

```python
mart = pd.read_excel(r"supermarket_sales.xlsx")
mart = mart[['Gender','Payment','Unit price','Quantity','Total','gross income']]
mart.head(10)
```

### 1. Create a basic KDE plot for Total column

```python
sns.kdeplot(data=mart, x='Total')
```

### 2. Create KDE for all the numeric variables in dataframe

```python
sns.kdeplot(data=mart)
```

### 3. Adjust the smooting using bw_adjust

```python
sns.kdeplot(data=mart, x='Total', bw_adjust=0.5)
```

### 4. Group the KDE on a category variable

```python
sns.kdeplot(data=mart,x='Total',hue='Payment')
```

### 5. Stack KDE on a category using MULTIPLE arguement

```python
sns.kdeplot(data=mart, x='Total', hue='Payment', multiple='stack')
```

### 6. Use log scaling to map the variable in KDE

```python
sns.kdeplot(data=mart, x='Total', log_scale=True)
```

### 7. Change styling of hued KDE using linewidth, palette, alpha etc...

```python
sns.kdeplot(data=mart,x='Total',hue='Payment',multiple='stack',linewidth=5,palette='Dark2',alpha=0.5)
```

### 8. Creat a bivariate KDE

```python
sns.kdeplot(data=mart, x='Unit price', y='gross income')
```

### 9. Group the bivariate KDE on a categorical variable and show the contours

```python
sns.kdeplot(data=mart, x='Unit price', y='gross income', hue='Gender', fill=True, levels=5, thresh=0.2)
```

### 5. Combine this with a KDE Plot

```python
plt.figure(figsize=(15,5))
sns.kdeplot(data=mart, x='Unit price', y='gross income',  hue='Gender', fill=True)
sns.rugplot(data=mart, x='Unit price', y='gross income', hue='Gender')
```

### 6. Combine this with a SCATTER PLOT

```python
sns.scatterplot(data=mart, x='Unit price',y='gross income',hue='Gender')
sns.rugplot(data=mart, x='Unit price',y='gross income',hue='Gender')
```

### 7. Show Rugs outside of the Axis

```python
plt.figure(figsize=(15,5))
sns.kdeplot(data=mart, x='Unit price', y='gross income', hue='Gender', fill=True)
sns.rugplot(data=mart, x='Unit price', y='gross income', hue='Gender', height=-0.01, clip_on=False)
```
