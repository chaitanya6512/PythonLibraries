### What is lineplot?
- It is used to draw line plots that visualize the relationship between two continues variables, x and y, with the possibility of incorporating additional semantic grouping through the hue, size, and style parameters.

### Syntax - 
- sns.lineplot(data=None, *, x=None, y=None, hue=None, size=None, style=None, units=None, weights=None, palette=None, hue_order=None, hue_norm=None, sizes=None, size_order=None, size_norm=None, dashes=True, markers=None, style_order=None, estimator='mean', errorbar=('ci',95), n_boot=1000, seed=None, orient='x', sort=True, err_style='band', err_kws=None, legend='auto', ci='deprecated', ax=None, **kwargs)

### Imported required libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Import Dataset

```python
mart = pd.read_csv(r'D:\Learnerea\Tables\train.csv')
mart[['Tier']]=mart['Outlet_Location_Type'].str.split(expand=True)[1]
mart.head()
```

### 1. Creat a basic line plot and try different estimators

```python
height = [5, 6, 7, 6.5,4]
age    = [22,25,30,25,25]
sns.lineplot(x=age, y=height, estimator=sum)
```

```python
sns.lineplot(data=mart,x="Outlet_Year",y="Sales",ci=None,estimator=None)
```

### 2. Test with different Confidence Intervals and with different number of Bootstraping

```python
sns.lineplot(data=mart, x="Outlet_Year", y="Sales", n_boot=2)
```

### 3. Grouping using HUE and use different Palettes

```python
sns.lineplot(data=mart, x="Outlet_Year", y="Sales", ci=None
             , hue='Outlet_Size'
            , palette=['red','green','blue'])
```

### 5. Grouping using Size

```python
sns.lineplot(data=mart, x="Outlet_Year", y="Sales", ci=None
            , size='Outlet_Size'
            , sizes=(.5,5))
```
