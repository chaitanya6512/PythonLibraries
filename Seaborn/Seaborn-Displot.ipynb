### What is displot?
- This function provides access to several approches for visualizing the univariate (or) bivariate distribution of data, incliuding subsets of data defined by semantic mapping and faceting across multiple subplots.
- The kind parameter selects the approch to use -
-   1. histplot() - with kind='hist'; Default.
-   2. kdeplot() - with kind='kde'.
-   3. ecdfplot() - with kind='ecdf'; univariate-only
 


### Syntax 
- sns.displot(data=None, *,x=None, y=None, hue=None, row=None, col=None, weights=None, kind='hist', rug=False, rug_kws=None, legend=True, palette=None, hue_order=None, color=None, height=5, **kwargs)

### Import required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Import Dataset

```python
mart = pd.read_excel(r"supermarket_sales.xlsx")
mart = mart[['Gender','Payment','Unit price','Quantity','Total','gross income','Branch']]
mart.head()
```

### displot()

```python
sns.displot(data=mart, y='Total', kde=True, rug=True, rug_kws={'height':0.05}, 
            hue='Payment', multiple='stack', col='Branch', row='Gender')
```

```python
sns.displot(data=mart, y='Total', kind='ecdf', hue='Payment', col='Branch', row='Gender')
```
