### What is ECDF plot?
- ECDF - Empirical cumulative distribution
- Which visualize the proportion (or) count of observations falling below each unique value in a dataset.
- Syntax - sns.ecdfplot(data=None, *, x=None, y=None, hue=None, weights=None, stat='Proportion', complementary=False, palette=None, hue_order=None, hue_norm=None, log_scale=None, legend=True, ax=None, **kwargs)
