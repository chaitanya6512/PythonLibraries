### what is jointplot?
- Draw a plot of two variables with bivariate and univariate graphs.This function provides a convenient interface to the 'JointGrid' class, with several canned plot kinds.
- This is intended to be a fairly light weight wrapper; if you need more flexibility, you should use : class:'JointGrid' directly.
  

### Syntax 
- sns.jointplot(data=None, *, x=None, y=None, hue=None, kind='scatter', height=6, ratio=5, space=0.2, dropna=False, xlim=None, ylim=None, color=None, palette=None, hue_order=None, hue_norm=None, marginal_ticks=False, joint_kws=None, mrginal_kws=None, **kwargs)

### Imported required libraries

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

### Import Dataset

```python
iris = sns.load_dataset('iris')
iris.head()
```

### 1. Drawing a first basic JOINT Plot

```python
sns.jointplot(data=iris, x='sepal_length', y='petal_length')
```

### 2. Change its kind to “scatter”|“kde”|“hist”|“hex”|“reg”|“resid”

```python
sns.jointplot(data=iris, x='sepal_length', y='petal_length'
             , kind='resid')
```

### 3. Grouping based on a categorical variable

```python
sns.jointplot(data=iris, x='sepal_length', y='petal_length'
             , hue='species')
```

### 4. Formatting All and Individual Plots

```python
sns.jointplot(data=iris, x='sepal_length', y='petal_length'
             , joint_kws=dict(marker='+'
                             ,color='red')
              ,marginal_kws=dict(color='green', kde=True))
```

### 5. Adjust height, ratio, space and show/hide the marginal_ticks

```python
sns.jointplot(data=iris,x='sepal_length',y='petal_length'
             ,height=5
             ,ratio=3
              ,space=5
             ,marginal_ticks=True)
```

### 6. Plot KDE and RUG on top of Joint Plot

```python
l=sns.jointplot(data=iris,x='sepal_length',y='petal_length',joint_kws=dict(color='red'))
l.plot_joint(sns.kdeplot)
l.plot_joint(sns.rugplot,height=0.1,color='red')
```
