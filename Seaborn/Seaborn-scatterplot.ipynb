### What is scatterplot?
- It is fundamental tool for visualizing the relationship between two numerical variables.
- Displaying data points on a two-Dimensional plane where each point represents an observation from a dataset.
  

### Syntax 
- sns.scatterplot(data=None, *, x=None, y=None, hue=None, size=None, style=None, palette=None, hue_order=None, hue_norm=None, sizes=None, size_order=None, size_norm=None, markers=True, style_order=None, legend='auto', ax=None, **kwargs)

### Imported required libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Import Dataset

```python

mart=pd.read_excel(r'mart_train_sample.xlsx')
mart.head()
```

### 1. Create a basic SCATTER PLOT

```python
sns.scatterplot(data=mart, x='Item_MRP', y='Sales')
```

### 2. Grouping basis on a categorical variable using HUE

```python
sns.scatterplot(data=mart, x='Item_MRP', y='Sales', hue='Outlet_Size')
```

### 3. Grouping basis on a categorical variable using STYLE and Styling Markers as well

```python
sns.scatterplot(data=mart, x='Item_MRP', y='Sales', style='Outlet_Size'
                , markers={"High":"^","Small":"v","Medium":"o"})
```

### 4. Grouping basis on a categorical variable using HUE & STYLE both together

```python
sns.scatterplot(data=mart, x='Item_MRP', y='Sales', style='Outlet_Size'
                , hue='Outlet_Size'
                , s=200
                , markers={"High":"^","Small":"v","Medium":"o"})
```

### 5. Grouping basis on numeric variable using HUE and use PALETTE

```python
sns.scatterplot(data=mart,x='Item_MRP',y='Sales'
#                 ,style='Outlet_Size'
                ,hue='Outlet_Size'
                ,palette='Purples_r'
                ,markers={"High":"^","Small":"v","Medium":"o"})
```

### 6. Grouping basis on a numeric variable using HUE & SIZE

```python
sns.scatterplot(data=mart, x='Item_MRP', y='Sales'
                , hue='Outlet_Size'
                , size='Tier'
                , sizes=(20,200)
                , legend='full'
                , palette='Purples_r'
                , markers={"High":"^","Small":"v","Medium":"o"})
```

### 7. Change the Marker size with S arguement

```python
sns.scatterplot(data=mart, x='Item_MRP', y='Sales', s=500, color='red', edgecolor='black')
```
