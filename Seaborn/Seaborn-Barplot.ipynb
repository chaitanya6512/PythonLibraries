### Seaborn.barplot()
- It is used for aggregate the categorical data according to some methods and default it's by mean.
- Syntax - sns.barplot( data, x=None, y=None, hue=None, order=None, hue_order=None, estimator='mean', errorbar=('ci', 95), n_boot=1000, seed=None, units=None, weights=None, orient=None, color=None, palette=None, saturation=0.75, fill=True, hue_norm=None, width=0.8, dodge='auto', gap=0, log_scale=None, native_scale=False, formatter=None, legend='auto', capsize=0, err_kws=None, ci=<deprecated>, errcolor=<deprecated>, errwidth=<deprecated>, ax=None, **kwargs).

### Import all the required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Display the SEABORN dataset names and load one of them for example.

```python
sns.get_dataset_names()
titanic = sns.load_dataset('titanic')
titanic.head()
```

###  Import mart dataframe

```python
mart = pd.read_excel('supermarket_sales.xlsx')
mart.head()
```

### Create a basic BARPLOT

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Product line', y='Total', data=mart)
```

### Add HUE to the barplot

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Product line',y='Total',hue='Gender',data=mart)
```

### Make the barplot HORIZONTAL

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Total',y='Product line',data=mart)
```

###  Plot the bars in a given ORDER

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Product line', y='Total', hue='Gender', data=mart, order=['Electronic_accessories',
```

### Add CAP on the ERROR BAR

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Product line',y='Total',hue='Gender',data=mart,order=['Electronic_accessories', 'Fashion_accessories','Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle','Sports_and_travel']
           ,hue_order=['Male','Female']
           ,capsize=0.2)
```

### REMOVE the error bar using ci

```python
plt.figure(figsize=(15,5))

sns.barplot(x='Product line',y='Total',hue='Gender',data=mart,order=['Electronic_accessories', 'Fashion_accessories','Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle','Sports_and_travel']
           ,hue_order=['Male','Female']
           ,ci=None, errorbar=None)
```

### Change bar colors using COLOR attribute

```python
plt.figure(figsize=(15,5))

sns.barplot(x='Product line',y='Total',hue='Gender',data=mart,order=['Electronic_accessories', 'Fashion_accessories','Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle','Sports_and_travel']
           ,hue_order=['Male','Female']
           ,ci=None, color='red')
```

### Change color using PALLETE attribute

```python
plt.figure(figsize=(15,5))

sns.barplot(x='Product line',y='Total',hue='Gender',data=mart,order=['Electronic_accessories', 'Fashion_accessories','Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle','Sports_and_travel']
           ,palette='cividis')
```

### Using SATURATION parameter

```python
plt.figure(figsize=(15,5))

sns.barplot(x='Product line',y='Total',hue='Gender',data=mart,order=['Electronic_accessories', 'Fashion_accessories','Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle','Sports_and_travel']
           ,palette='cividis', saturation=0.2)
```

### Change default parameter by using ESTIMATER.

```python
plt.figure(figsize=(15,5))

sns.barplot(x='Product line',y='Total',hue='Gender',data=mart,order=['Electronic_accessories', 'Fashion_accessories','Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle','Sports_and_travel']
           ,palette='cividis', estimator=np.median)
```
