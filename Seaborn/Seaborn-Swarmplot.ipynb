### What is swarmplot
- It is similar to stripplot.Only points to be adjusted it won't be overlap as it helps to represent better representation of distribution of values.
- Syntax - sns.swarmplot(data=None, *, x=None, y=None, hue=None, order=None, hue_order=None, dodge=False, orient=None, color=None, palette=None, size=5, edgecolor=None, linewidth=0, hue_norm=None, log_scale=None, native_scale=False, formatter=None, legend='auto', warn_thresh=0.05, ax=None, **kwargs)

### 1. Import all the required libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### 2. Import Dataset

```python
mart = pd.read_excel('supermarket_sales.xlsx')
mart.head()
```

### 3. Create a basic SWARM PLOT on Total column

```python
sns.swarmplot(y='Total', data=mart)
```

### 4. Create SWARM PLOT group by CATEGORIES

```python
sns.swarmplot(x='Payment', y='Total', hue='Gender', data=mart)
```

### 5. Showing Hues Separatly on the categorical axis

```python
plt.figure(figsize=(10,5))
sns.swarmplot(x='Payment', y='Total', hue='Gender', data=mart)
```

### 6. Styling a SWARM - Change the marker, size, color, edge color.....

```python
plt.figure(figsize=(15,5))
sns.swarmplot(x='Payment', y='Total', hue='Gender', data=mart, size=5, palette='dark:green', edgecolor='black')
```

### 7. Overlay a SWARM PLOT on a BOX PLOT

```python
sns.swarmplot(x='Payment', y='Total', data=mart, edgecolor='black')
sns.boxplot(x='Payment', y='Total', data=mart)
```

### 8. Overlay a SWARM PLOT on a VIOLIN PLOT

```python
sns.swarmplot(x='Payment', y='Total', data=mart, palette='dark:white')
sns.violinplot(x='Payment', y='total', data=mart, inner=None, width=1)
```
