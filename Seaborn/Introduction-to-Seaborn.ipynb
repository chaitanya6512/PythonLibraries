### What is Seaborn ?
- Seaborn is a python data visualization library based on matplotlib.
- It provides high-level interface for drawing attractive and informative  statistical graphics.
#### Seaborn Plots:
- 1. Categorical -
     - Countplot()
     - Barplot()
     - Boxplot()
     - Violinplot()
     - Stripplot()
     - Swarmplot()
     - Factorplot()
- 2. Distribution -
     - Displot()
     - Rugplot()
     - Pairplot()
     - Joinplot()
- 3. Relation -
- 4. Regression -
     - Simple linear plot()
     - Linear plot with additional parameters.
     - Setting the size and color of the plot.
     - Displaying multiple plots.
     - Size and aspect ratio of theplots.
- 5. Matrix -
     - Heat maps
     - Cluster maps
- 6. Multi-plot grids
