### Importing all the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Import the data

```python
mart = pd.read_excel('supermarket_sales.xlsx')
mart.head()
```

### Create a COUNTPLOT to get number of transactions for each of the product line and change the figure size.
- sns.countplot(data=None, *, x=None, y=None, hue=None, order=None, hue_order=None, orient=None, color=None, palette=None, saturation=0.75, fill=True, hue_norm=None, stat='count', width=0.8, dodge='auto', gap=0, log_scale=None, native_scale=False, formatter=None, legend='auto', ax=None, **kwargs)
- Paramerters are : -
   - data - DataFrame, Series, dict, array, (or) list of arrays.
   - x, y, hue - name of variables in data (or) vector data.
   - orient - Orientation of the plot ( vertical (or) horizontal ).
   - color - single color for the elements in the plots.
   - palette - palette name, list, (or) dict. Colors to use for the different levels of the hue variable.
   - saturation - float.
   - hue_norm - tuple (or) matplotlib.colors.Normalize Object.
   - stat - {'count', 'percent', 'proportion', 'probability'}.
   - width - float.
   - log_scale - bool (or) number, or pair of bools (or) numbers.
   - native_scale - bool.
   - formatter - Converting categorical datainto strings.
   - legend - "auto", "brief", "full", (or) False.
   - kwargs - key, value mappings.

```python
plt.figure(figsize=(15,5))
sns.countplot(x='Product line', data=mart)
```

#### Make it horizontal

```python
sns.countplot(y='Product line', data=mart)
```

#### Add hue to get the count on two categories i.e. Product line and Gender

```python
plt.figure(figsize=(15,5))
sns.countplot(x='Product line', hue='Gender', data=mart)
```

#### Using different color palette

```python
sns.countplot(x="Product line", hue="Gender", data=mart, palette='PuBuGn_r')
```

#### Change Style using facecolor, linewidth and edge color

```python
plt.figure(figsize=(15,5))

sns.countplot(x='Product line',data=mart
             ,facecolor = (0,0,0,0)
             ,linewidth=5
             ,edgecolor=sns.color_palette('dark',3))
```
