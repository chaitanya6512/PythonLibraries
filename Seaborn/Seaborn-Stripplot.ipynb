### What is stripplot?
- It is a categorical scatterplot used to visualize the distribution of a continuous variable across different categories.
- Syntax - sns.stripplot(data=None, *, x=None, y=None, order=None, hue_order=None, jitter=True, dodge=False, orient=None, color=None, palette=None, size=5, edgecolor=<default>, linewidth=0, hue_norm=None, log_scale=None, native_scale=False, formatter=None, legend='auto', ax=None, **kwargs)

### Import all the required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Import DataFrame

```python
mart = pd.read_excel("supermarket_sales.xlsx")
mart.head()
```

### Create a basic STRIP PLOT

```python
sns.stripplot(y='Total', data=mart)
```

### Expand markers in STRIP PLOT using jitter

```python
sns.stripplot(y='Total', data=mart, jitter=0.5)
```

### Draw line around the points using LINEWIDTH

```python
sns.stripplot(y="Total", data=mart, linewidth=0)
```

### Include third categorical variable with HUE

```python
sns.stripplot(x='Payment', y='Total', data=mart, hue='Gender', jitter=0.3, linewidth=0.5)
```

### Separate each level of hue using DODGE

```python
sns.stripplot(x='Payment', y='Total', data=mart, hue='Gender',jitter=0.2, linewidth=0.5, dodge=True)
```

### Draw the strips on top of VIOLIN PLOT

```python
sns.stripplot(data=mart, y='Total',x='Payment',jitter=0.1)
sns.violinplot(data=mart, y='Total',x='Payment',color='gray')
```

### Draw the strips on top of BOX PLOT

```python
sns.stripplot(x='Payment', y='Total', data=mart, hue='Gender', jitter=0.2)
sns.violinplot(x='Payment', y='Total', data=mart, hue='Gender', palette='gray')
```
