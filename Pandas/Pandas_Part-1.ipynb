# Data Cleaning
- Data Cleaning means fixing bad data in your data set.
- Bad Data Could be : Duplicates, Wrong data,data in wrong format, Empty cells.

# Pandas - OneWay Cleaning Empty cells
- Empty cells : It could be give you a wrong result when analyze the wrong data.
- One way To deal with empty rows : To remove empty rows that contain empty cells.
- dropna() : Which is used for removes rows whith missing values.
- By default,dropna() : returns a newDataframe and will not change original.
- In case, if we want to change the original Dataframe,we use "inplace = True" argument.
- dropna(inplace = True) : Return original dataframe and remove empty rows in the original dataset and not return the new dataframe. 

```python
import pandas as pd
df=pd.read_csv('data.csv')
new_df=df.dropna()
print(new_df.to_string())


import pandas as pd
df=pd.read_csv('data.csv')
df.dropna(inplace=True)
print(df.to_string())
```

# Another way to Cleaning Empty Cells
- Another way of dealing with empty cells to insert new value instead.
- fillna() : This method is used to replace empty values with a new value.
- we can also replace for one specific column by specifying a one.

```python
import pandas as pd
df=pd.read_csv('data.csv')
df.fillna(130, inplace=True)
print(df.to_string())

import pandas as pd
df=pd.read_csv('data.csv')
df.fillna({"Calories":130},inplace=True)
print(df.to_string())
```

# Replace Using Mean, Median, or Mode
- A common way to replace empty cells, is to calculate the mean, median or mode value of the column.
- Mean() : The average value (the sum of all values divided by number of values).
- Median() : The value in the middle, after you have sorted all values ascending.
- Mode() : The value that appears most frequently.

```python
import pandas as pd
df=pd.read_csv('data.csv')
x=df['Calories'].mean()
df.fillna({'Calories':x},inplace=True)
print(df.to_string())


import pandas as pd
df=pd.read_csv('data.csv')
y=df['Calories'].median()
df.fillna({'Calories':y},inplace=True)
print(df.to_string())


import pandas as pd
df=pd.read_csv('data.csv')
z=df['Calories'].mode()[0]
df.fillna({'Calories':z},inplace=True)
print(df.to_string())
```

# Pandas - Cleaning Data of Wrong Format
- Cells with data of wrong format can make it difficult or even impossible to analyze it.
- To fix it, you have two solutions : remove the rows , or convert all cells in the columns in the same format.

# One way : Remove rows
- dropna() : Removes rowns inthe dataset.

```python
import pandas as pd
df = pd.read_csv('data.csv')
df['Date'] = pd.to_datetime(df['Date'], format='mixed')
df.dropna(subset=['Date'], inplace = True)
print(df.to_string())
```

# Another Way : convert all cells into same format
- In this Date Column,we can format by using to_datetime() method.

```python
import pandas as pd
df = pd.read_csv('data.csv')
df['Date'] = pd.to_datetime(df['Date'], format='mixed')
print(df.to_string())
```

```python
# Pandas - Fixing Wrong Data
- "Wrong data" does not have to be "empty cells" or "wrong format", it can just be wrong, like if someone registered "199" instead of "1.99".

# one way to fix wrong data
- One way to fix wrong values is to replace them with something else.
- For small data sets you might be able to replace the wrong data one by one, but not for big data sets.
- In our example, it is most likely a typo, and the value should be "45" instead of "450", and we could just insert "45" in row 7:
```

```python
import pandas as pd
df = pd.read_csv('data.csv')
df.loc[7,'Duration'] = 45
print(df.to_string())

import pandas as pd
df = pd.read_csv('data.csv')
for x in df.index:
  if df.loc[x, "Duration"] > 120:
    df.loc[x, "Duration"] = 120
print(df.to_string())
```

# Another way to fix wrong data
- Another way of handling wrong data is to remove the rows that contains wrong data.

```python
import pandas as pd
df = pd.read_csv('data.csv')
for x in df.index:
  if df.loc[x, "Duration"] > 120:
    df.drop(x, inplace = True)
print(df.to_string())
```

# Pandas - Removing Duplicates
## To discover duplicate:
- duplicated() - To discover duplicates, we can use this method.
- This method returns a duplicated values in boolean values in each row.

## To remove duplicates:
- drop_duplicates() - To remove duplicates.

```python
import pandas as pd
df=pd.read_csv('data.csv')
print(df.duplicated())


import pandas as pd
df=pd.read_csv('data.csv')
df.drop_duplicates(inplace=True)
print(df.to_string())
```

# Pandas - Data Correlations
## Finding Relationships
- corr() - method calculates the relationship between each column in your dataset.
- The number varies from -1 to 1.
- 1 - means there is a 1 to 1 realtionship(a perfect relationship).
- 0.9 - is also a good relationship as 0.9,if you increase one value ,other probably increase as well.
- (-0.9)- would be just as good relationship as 0.9 ,if you increase one value then other probably go down.
-  0.2 means NOT a good relationship, meaning that if one value goes up does not mean that the other will.

## What is a good correlation? 
- It depends on the use, but I think it is safe to say you have to have to have at least 0.6 (or -0.6) to call it a good correlation.

```python
import pandas as pd
df=pd.read_csv('data.csv')
print(df.corr())
```
