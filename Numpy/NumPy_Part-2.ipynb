### Random Numbers in NumPy
- Random Number does Not mean a different number everytime.Random means something that can not be predicted logically.
- random() - NumPy offers a random module to work with random numbers.
- randint(100) - Return random integer values from 1 to 100.
- rand() - Returns a random float between 0 to 1.
- size() - To specify shape of the array.
- choice() - It allows a random value based on an array values. 

```python
from numpy import random
x = random.randint(100)
print(x)

from numpy import random
x = random.rand()
print(x)

from numpy import random
x = random.rand(5)
print(x)

from numpy import random
x = random.randint(100, size=(5))
print(x)

from numpy import random
x = random.randint(100, size=(3,5))
print(x)

from numpy import random
x = random.rand(10)
print(x)

from numpy import random
x = random.rand(3,5)
print(x)

from numpy import random
x = random.choice([1,4,7,5])
print(x)

from numpy import random
x = random.choice([1,4,7,5], size=(3,5))
print(x)
```

### NumPy Random Data Distribution
- Data Distribution - It is alist of all possible values, and how often each value occurs.
### Random Distribution 
- It is aset of random numbers that follow a certain probability density function.
- Probability density function - A function that describes a continuous probability.
- The Sum of all probability numbers must be 1.
  

```python
from numpy import random
x = random.choice([3,5,7,9], p=[0.1,0.3,0.6,0.0], size=(10))
print(x)

from numpy import random
x = random.choice([3,5,7,9], p=[0.1,0.3,0.6,0.0], size=(3,5))
print(x)
```

### Random Permutations
- A Permutations refers to arrangement of elements.
- NumPy Random module provides two methods :
- 1. shuffle() - Make changes to the original array.
- 2. permutation() - Method returns a re-arranged array(and leaves the original array un-changed).
     

```python
from numpy import random
import numpy as np
x = np.array([1,4,7,2,5,8])
random.shuffle(x)
print(x)

from numpy import random
import numpy as np
x = np.array([1,4,7,2,5,8])
random.permutation(x)
print(x)
```

### Normal Distribution (or) Gaussian Distribution
- It is a continous Distribution.
- It fits probaility distribution for many events.
- random.normal() - To get normal Data Distribution.
- It has three attributes :
- 1. loc - mean (Where peak bell value exists).
  2. scale - Standard distribution
  3. size - shape of the returned array.
- The Normal Distribution is also called as Bell Curve because of bell-shaped curve.

```python
from numpy import random
x = random.normal(size=(2,3))
print(x)

from numpy import random
x = random.normal(loc= 1, scale=2, size=(2,3))
print(x)

from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
sns.displot(random.normal(size=1000),kind="kde")
plt.show()
```

### Poisson Distribution
- It is a Discrete Distribution.
- It estimates how many times that event can be happen in a specificed time.
- Eg : If you twice a day what is the probability of thrice a day ?
- It takes two parameters :
- lam : rate (or) known number of occurances.
- size : Shape of the array.

```python
from numpy import random
x = random.poisson(lam=2, size=10)
print(x)

from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
sns.displot(random.poisson(lam=2, size=10))
plt.show()

from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
sns.displot(random.poisson(lam=2, size=10), kind="kde")
plt.show()
```

### Zipf Distribution
- Zipf distributions are used to sample data based on zipf's law.
- zipf's law - In a collection, the nth common term is 1/n times of the most common term.
- Eg : The 5th most common word in English occurs nearly 1/5 times as often as the most common word.
- It has two parameters :
- a - distribution parameter.
- size - shape of the parameter.

```python
from numpy import random
x = random.zipf(a=2, size=(2, 3))
print(x)

from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
x = random.zipf(a=2, size=1000)
sns.displot(x[x<10])
plt.show()

from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
x = random.zipf(a=2, size=1000)
sns.displot(x[x<10],kind="kde")
plt.show()
```
